```python
regressor = Sequential()


regressor.add(Dense(units=300, kernel_regularizer="l1", bias_regularizer='l1', kernel_initializer = 'uniform',activation='relu',input_dim =X_train.shape[1]))
regressor.add(BatchNormalization())
regressor.add(Dense(units=150,kernel_regularizer="l1", bias_regularizer='l1',kernel_initializer="uniform",activation='relu'))
regressor.add(BatchNormalization())
regressor.add(Dense(units=75,kernel_regularizer="l1", bias_regularizer='l1',kernel_initializer="uniform",activation='relu'))
regressor.add(BatchNormalization())

regressor.add(Dense(units=50, kernel_regularizer='l1', bias_regularizer='l1',kernel_initializer = 'uniform', activation = 'relu'))
regressor.add(BatchNormalization())
regressor.add(Dense(units=25, kernel_regularizer='l1', bias_regularizer='l1', kernel_initializer = 'uniform', activation = 'relu'))
regressor.add(BatchNormalization())
regressor.add(Dense(units=15, kernel_regularizer='l1', bias_regularizer='l1', kernel_initializer = 'uniform', activation = 'relu'))
regressor.add(BatchNormalization())
regressor.add(Dense(units=8, kernel_regularizer='l1', bias_regularizer='l1',kernel_initializer = 'uniform', activation = 'relu'))
regressor.add(BatchNormalization())
regressor.add(Dense(units=4, kernel_regularizer='l1', bias_regularizer='l1', kernel_initializer = 'uniform', activation = 'relu'))
regressor.add(BatchNormalization())
regressor.add(Dense(units=2, kernel_regularizer='l1', bias_regularizer='l1', kernel_initializer = 'uniform', activation = 'relu'))
regressor.add(BatchNormalization())
regressor.add(Dense(units=1, kernel_regularizer='l1', bias_regularizer='l1',kernel_initializer = 'uniform', activation = 'linear'))
regressor.add(BatchNormalization())
regressor.compile(optimizer = 'Adamax', loss = 'mean_squared_error', metrics = ['mse', 'mae', 'mape'])
model_history=regressor.fit(X_train, y_train,validation_split=0.30, batch_size = 1500, epochs=300)
```

```python
y_pred = regressor.predict(X_test)
```
